```python
import pm4py

log1 = pm4py.read_xes('1167521718.xes')
pd1 = pm4py.convert_to_dataframe(log1)
pd1
```

```python
import pm4py

log2 = pm4py.read_xes('1167521718_noisy.xes')
pd2 = pm4py.convert_to_dataframe(log2)
pd2
```

```python
# combine the rows by case number 
pd1 =pd1.groupby('ID', as_index=False).agg(lambda x: x.tolist())
pd1
```

```python
# combine the rows by case number 
pd2 =pd2.groupby('ID', as_index=False).agg(lambda x: x.tolist())
pd2
```

```python
pd2['concept:name_normal'] = pd1['concept:name'] 
```

```python
pd2
```

```python
pd=pd2[['ID','concept:name','concept:name_normal']]
pd
```

```python
import numpy as np
pd['Status'] = np.where(pd['concept:name'] == pd['concept:name_normal'], 'Normal', 'Anomalous')
pd
```
